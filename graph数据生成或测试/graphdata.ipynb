```python
person_df= pd.read_csv(base_dir+file_names[1],encoding='utf-8')
person_df[person_df.columns[0]] = person_df[person_df.columns[0]]-243000
base_id = list(person_df[person_df.columns[0]])
person_df[person_df.columns[0]] = 'person_'+ person_df[person_df.columns[0]].astype('str')
person_df['name'] = [('name_'+ str(x)) for x in base_id]
person_df
```

```python
co = person_df.columns    
person_ID = list(person_df[co[0]])
name = list(person_df[co[1]])
address = list(person_df[co[2]])
age = list(person_df[co[3]])
flag = list(person_df[co[4]])
status = list(person_df[co[5]])
for i in range(1,10000):
    person_ID = person_ID + [('person_'+ str(x+i*20)) for x in base_id]
    name = name + [('name_'+str(x+i*20)) for x in base_id]
    address = address + list(person_df[co[2]])
    age = age +[(x+random.randint(-6,14)) for x in person_df[co[3]]]
#     age = age + list(person_df[co[3]])
    flag = flag + list(person_df[co[4]])
    status = status + list(person_df[co[5]])
print('=====END=====')
person = pd.DataFrame()
person["personID:ID(personID)"] = person_ID
person['name'] = name
person['address'] = address
person['age'] = age
person['flag'] = flag
person['status'] = status
person[':LABEL'] = 'person'
person
```

```python
person.to_csv(r'out/person.csv', encoding='utf-8', index=False)
```
